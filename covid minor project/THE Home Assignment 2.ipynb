```python
root = Tk()
root.title("Pandemic Analytics Engine")

root.configure(background = "light grey")

root.geometry("300x600")

lbl=Label(root,text="Pandemic Analyzer", fg = "blue",font=("SUNKEN",30))
lbl.grid(row=1,column=1)
```

```python
def calculate():

    population = e2.get()
    int_total_ppl = int(population)

    infected = e3.get()
    int_infected = int(infected)

    deceased = e4.get()
    int_deceased = int(deceased)

    ifr = (int_deceased / int_infected) * 100
    cmr = (int_deceased / int_total_ppl)*100
    
    

    lbl1 = Label(root, font=("SUNKEN", 22, "bold"), text="value")
    lbl1.grid(row=7, column=1)

    lbl2 = Label(root, font=("SUNKEN", 12, "bold"), text="IFR : ")
    lbl2.grid(row=9, column=1)
    
   

    lbl3 = Label(root, text=ifr)
    lbl3.grid(row=9, column=2)

    lbl4 = Label(root, font=("SUNKEN", 12, "bold"), text="CMR : ")
    lbl4.grid(row=10, column=1)
   
   
    lbl5 = Label(root, text=cmr)
    lbl5.grid(row=10, column=2)

    
options = ["Maharashtra","Tamil Nadu","Delhi","Gujarat","Uttar Pradesh","West Bengal","Rajasthan","Telengana","Karnataka","Andhra Pradesh","Haryana","Madhya Pradesh","Bihar","Assam","Jammu and Kashmir","Odisha","Punjab","Kerala","Chhattisgarh","Uttarakhand","Jharkhand","Goa","Tripura","Manipur","Himachal Pradesh","Ladakh","Puducherry","Nagaland","Chandigarh","Dadra and Nagar Haveli","Arunachal Pradesh","Mizoram","Andaman and Nicobar","Sikkim","Meghalaya","Lakshadweep"]

select = StringVar() 
select.set(options[0])
```

```python
lbl2 = Label(root, text="Select State", fg="white")
lbl2.grid(row=2)

drop = OptionMenu(root,select,*options)
drop.grid(row=2,column=2)

lbl3 = Label(root, text="Total number of Population", fg="purple")
lbl3.grid(row=3)

e2 = Entry(root)
e2.grid(row=3, column=2)

lbl4 = Label(root, text="Infected",fg="yellow")
lbl4.grid(row=4)
e3 = Entry(root)
e3.grid(row=4, column=2)



lbl5 = Label(root, text="Deceased" ,fg="red")
lbl5.grid(row=5)
e4 = Entry(root)
e4.grid(row=5, column=2)

lbl6 = Label(root, text="Recovered", fg="green")
lbl6.grid(row=6)







lbl7 = Label(root, text="Value",fg="blue")
lbl7.grid(row=12)


button1 = Button(root, text="Calculate IFR & CMR",command=calculate)
button1.grid(row = 13, column=2)


root.mainloop()
```
